```python
def relations(n, edges): # Возвращает список множеств соседей (индексация с 0)
    rel = [set() for _ in range(n)]
    for u, v in edges:
        u0 = u - 1
        v0 = v - 1
        rel[u0].add(v0)
        rel[v0].add(u0)
    return rel

def max_independent_set(rel): # Возвращает максимальное независимое множество
    n = len(rel)
    candidates = set(range(n))
    best = []

    def selection(cand, current):
        nonlocal best
        # Отсечение: даже если взять все оставшиеся, не превзойдём лучшее
        if len(current) + len(cand) <= len(best):
            return
        if not cand:
            if len(current) > len(best):
                best = current[:]
            return
        # Выбираем вершину с наибольшим количеством смежных вершин
        v = max(cand, key=lambda x: len(rel[x] & cand))
        # Ветвь 1: включаем v
        new_cand = cand - {v} - (rel[v] & cand)
        selection(new_cand, current + [v])
        # Ветвь 2: не включаем v
        new_cand = cand - {v}
        selection(new_cand, current)

    selection(candidates, [])
    return best
```
